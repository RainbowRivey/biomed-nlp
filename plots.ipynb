```python
import pandas as pd
model = 'tmvar'
path = f'_{model}'
df = pd.read_json(f"{path}/log_history.json")


df = df[df.eval_runtime.notnull()]
loss_fig = df.plot(x='epoch', y=['eval_loss'], kind='bar', title=f"{model} loss")
loss_fig.figure.savefig(f'{path}/loss.png')

eval_fig = df.plot(x='epoch', y=['eval_precision', 'eval_recall',
                                 'eval_f1'], kind='bar', figsize=(15, 9), title =f"{model} eval")
eval_fig.figure.savefig(f'{path}/eval.png')
```
